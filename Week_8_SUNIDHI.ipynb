# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

def agent(query: str):
    query_lower = query.lower()

    # Route 1 : Calculator
    if "calculate" in query_lower:

        expression = query_lower.replace("calculate", "").strip()

        result = calculator(expression)

        return {
            "type": "calculation",
            "result": result
        }

    # Route 2 : Keyword Extraction
    elif "keywords" in query_lower:

        text = query.replace("Extract keywords from", "").strip()

        result = extract_keywords(text)

        return {
            "type": "keywords",
            "result": result
        }

    # Route 3 : General Response
    else:

        return {
            "type": "general",
            "result": "I'm a simple smart assistant. I can perform calculations and extract keywords."
        }
````

## 📦 Expected Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# 🧪 Test Cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("Query:", q)
    print("Response:", agent(q))
    print("-" * 50)
````

````python
# 🎯 Interactive Mode

while True:
    user_input = input("Enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("Response:", agent(user_input))
````

ENHANCED AGENT

````python
#Adding a new tool - Word Counter

def word_counter(text: str):
    return len(text.split())
````

````python
# Enhanced Smart Agent

def enhanced_agent(query: str):

    query = query.strip()

    # Error Handling
    if not query:
        print("[LOG] Empty Query")
        return {
            "type": "error",
            "result": "Empty query received."
        }

    query_lower = query.lower()

    # ---------------- Calculator ----------------
    if any(word in query_lower for word in ["calculate", "compute", "solve"]):

        print("[LOG] Calculator Tool Selected")

        expression = query_lower

        for word in ["calculate", "compute", "solve"]:
            expression = expression.replace(word, "")

        expression = expression.strip()

        result = calculator(expression)

        return {
            "type": "calculation",
            "tool_used": "calculator",
            "result": result
        }

    # ---------------- Keyword Extraction ----------------
    elif any(word in query_lower for word in ["keyword", "keywords"]):

        print("[LOG] Keyword Tool Selected")

        text = query

        for word in [
            "Extract keywords from",
            "extract keywords from",
            "Extract keyword from",
            "extract keyword from",
            "keywords",
            "keyword"
        ]:
            text = text.replace(word, "")

        text = text.strip()

        result = extract_keywords(text)

        return {
            "type": "keywords",
            "tool_used": "keyword_extractor",
            "result": result
        }

    # ---------------- Word Counter ----------------
    elif "count words" in query_lower:

        print("[LOG] Word Counter Tool Selected")

        text = query_lower.replace("count words", "").strip()

        result = word_counter(text)

        return {
            "type": "word_count",
            "tool_used": "word_counter",
            "result": result
        }

    # ---------------- General ----------------
    else:

        print("[LOG] General Response")

        return {
            "type": "general",
            "tool_used": "none",
            "result": "I'm a Smart AI Assistant. I can calculate expressions, extract keywords and count words."
        }
````

````python
print("Enhanced Smart Assistant")
print("Type 'exit' to quit.")

while True:

    query = input("\nEnter your query: ")

    if query.lower() == "exit":
        print("Goodbye!")
        break

    response = enhanced_agent(query)

    print("Response:", response)
````
